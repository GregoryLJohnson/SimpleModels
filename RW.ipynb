```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the random walk
num_steps = 1000  # Number of steps in the walk

# Function to generate the random 2D walk
def random_walk_2d(n):
    # Generate random steps: -1, 0, or 1 for x and y directions
    steps_x = np.random.choice([-1, 0, 1], size=n)
    steps_y = np.random.choice([-1, 0, 1], size=n)
    
    # Cumulative sum to get positions
    x_positions = np.cumsum(steps_x)
    y_positions = np.cumsum(steps_y)
    
    return x_positions, y_positions

# Generate the walk
# x, y = random_walk_2d(num_steps)

# Plot the random walk
plt.figure(figsize=(8, 8))
for _ in range(10):
    x, y = random_walk_2d(num_steps)
    plt.plot(x, y, lw=1)
    plt.scatter(0, 0, c="green", label="Start", s=100)  # Starting point
    plt.scatter(x[-1], y[-1], c="red", label="End", s=100)  # Ending point
plt.title("Random 2D Walk")
plt.xlabel("X Position")
plt.ylabel("Y Position")
#plt.legend()
plt.grid()
plt.axis("equal")  # Ensure equal scaling for x and y axes
plt.show()
```
